```python
dataset = nltk.sent_tokenize(raw)
```

```python
#preprocessing buang karakter yg tidak perlu
for i in range (len(dataset)):
  dataset[i]=dataset[i].lower()
  #hapus non word dgn spasi
  dataset[i]=re.sub(r'\W',' ',dataset[i])
  dataset[i]=re.sub(r'\n',' ',dataset[i])
```

```python
#buat histogram , liat kemunculannya
wordToCount={}
for theSentences in dataset:
  theWords=nltk.word_tokenize(theSentences)
  for theWord in theWords:
    if theWord not in wordToCount.keys():
      wordToCount[theWord] = 1
    else:
      wordToCount[theWord] = 2
```

```python
# ambil 10 kata yang sering muncul
import heapq
wordFreq = heapq.nlargest(20,wordToCount,key=wordToCount.get)
wordFreq
```

```python
#hitung presentasi dalam vector
d=[]
for theData in dataset:
  vector=[]
  for words in wordFreq:
    if words in nltk.word_tokenize(theData):
      vector.append(1)
    else:
      vector.append(0)
  d.append(vector)
```

```python
#array 2d
d = np.asarray(d)
d
```

```python
#Probabilitas Kata
def readData():
  datas=newRaw
  data=[]
  for x in range(len(datas)):
    for word in datas[x].split():
      data.append(word)
  print(data)
  return data
```

```python
readData()
```

```python
#buat bigram
def createBigram(datas):
  listOfBigrams=[]
  bigramCounts={}
  unigramCounts={}
  for z in range(len(datas)-1):
    if z < len(datas) -1 and datas[z+1].islower():
      listOfBigrams.append((datas[z],datas[z+1]))
      if(datas[z],datas[z+1]) in bigramCounts:
        bigramCounts[(datas[z],datas[z+1])] +=1
      else:
        bigramCounts[(datas[z],datas[z+1])] = 1
    if datas[z] in unigramCounts:
      unigramCounts[datas[z]] +=1
    else:
      unigramCounts[datas[z]] =1
  return listOfBigrams,unigramCounts,bigramCounts
```

```python
w = readData()
a,b,c = createBigram(w)
```

```python
# probability masing masing bigram
def calcBigramProb(listOfBigrams,unigramCounts,bigramCounts):
  listOfProb={}
  for bigram in listOfBigrams:
    word1=bigram[0]
    word2=bigram[1]
    listOfProb[bigram]=(bigramCounts.get(bigram)/unigramCounts.get(word1))
  return listOfProb
```

```python
v = calcBigramProb(a,b,c)
```
